```python
from tkinter import ttk
from tkinter import messagebox
import sqlite3
from tkinter import *
```

```python
class Database:
    def __init__(self):
        self.con = sqlite3.connect("D:/Oracuz internship/dbFile1.db")
        self.cur = self.con.cursor()
        self.cur.execute("CREATE TABLE IF NOT EXISTS Medicine_info (id PRIMARYKEY text, Medicine_name text, mg text, Avilable_stock text, Expiry_date text)")

    def __del__(self):
        self.cur.close()
        self.con.close()

    def Insert(self, id,Medicine_name,mg,Avilable_stock,Expiry_date):
        self.cur.execute("INSERT INTO Medicine_info VALUES (?, ?, ?, ?, ?)", (id,Medicine_name,mg,Avilable_stock,Expiry_date))
        self.con.commit()
        
    def Update(self,mg,Avilable_stock,Expiry_date, id):
        self.cur.execute("UPDATE Medicine_info SET mg = ?,Avilable_stock = ?,Expiry_date = ? WHERE id = ?", (mg,Avilable_stock,Expiry_date, id))
        self.con.commit()
        
    def Search(self, id):
        self.cur.execute("SELECT * FROM Medicine_info WHERE id = ?", (id, ))
        searchResults = self.cur.fetchall()
        return searchResults
        
    def Delete(self, id):
        self.cur.execute("DELETE FROM Medicine_info WHERE id = ?", (id, ))
        self.con.commit()

    def Display(self):
        self.cur.execute("SELECT * FROM Medicine_info")
        records = self.cur.fetchall()
        return records

        
class InsertWindow:
    def __init__(self):
        self.window = Tk()
        self.window.title("Insert data")
        
        self.id = StringVar()
        self.Medicine_name = StringVar()
        self.mg = StringVar()
        self.Avilable_stock = StringVar()
        self.Expiry_date = StringVar()
        
        Label(self.window, text = "Enter details",  width = 25).grid(pady = 5, column = 2, row = 0)
        Label(self.window, text = "Medicine ID",  width = 25).grid(pady = 5, column = 1, row = 1)
        Label(self.window, text = "Medicine Name",  width = 25).grid(pady = 5, column = 1, row = 2)
        Label(self.window, text = "Milli gram",  width = 25).grid(pady = 5, column = 1, row = 3)
        Label(self.window, text = "Avilable Stock",  width = 25).grid(pady = 5, column = 1, row = 4)
        Label(self.window, text = "Expiry Date",  width = 25).grid(pady = 5, column = 1, row = 5)
        
        self.idEntry = Entry(self.window,  width = 25, textvariable = self.id)
        self.NameEntry = Entry(self.window,  width = 25, textvariable = self.Medicine_name)
        self.mgEntry = Entry(self.window,  width = 25, textvariable = self.mg)
        self.stockEntry = Entry(self.window,  width = 25, textvariable = self.Avilable_stock)
        self.expEntry = Entry(self.window,  width = 25, textvariable = self.Expiry_date)
        
        self.idEntry.grid(pady = 5, column = 3, row = 1)
        self.NameEntry.grid(pady = 5, column = 3, row = 2)
        self.mgEntry.grid(pady = 5, column = 3, row = 3)
        self.stockEntry.grid(pady = 5, column = 3, row = 4)
        self.expEntry.grid(pady = 5, column = 3, row = 5)

        Button(self.window, width = 20, text = "Insert",activebackground="grey", command = self.Insert).grid(pady = 15, padx = 5, column = 1, row = 14)
        Button(self.window, width = 20, text = "Close",activebackground="grey", command = self.window.destroy).grid(pady = 15, padx = 5, column = 3, row = 14)

        self.window.mainloop()

    def Insert(self):
        self.database = Database()        
        self.database.Insert(self.idEntry.get(), self.NameEntry.get(), self.mgEntry.get(), self.stockEntry.get(), self.expEntry.get())
        messagebox.showinfo("Inserted data", "Data inserted successfully")
    

class UpdateWindow:
    def __init__(self, id):
        self.window = Tk()
        self.window.title("Update data")
        
        self.id = id
        self.Medicine_name = StringVar()
        self.mg = StringVar()
        self.Avilable_stock = StringVar()
        self.Expiry_date = StringVar()

        Label(self.window, text = "Enter details",  width = 25).grid(pady = 5, column = 2, row = 0)
        Label(self.window, text = "Medicine ID",  width = 25).grid(pady = 5, column = 1, row = 1)
        Label(self.window, text = id,  width = 25).grid(pady = 5, column = 2, row = 1)
        Label(self.window, text = "Medicine Name",  width = 25).grid(pady = 5, column = 1, row = 2)
        Label(self.window, text = "Milli gram",  width = 25).grid(pady = 5, column = 1, row = 3)
        Label(self.window, text = "Avilable Stock",  width = 25).grid(pady = 5, column = 1, row = 4)
        Label(self.window, text = "Expiry Date",  width = 25).grid(pady = 5, column = 1, row = 5)
        
        # previous values
        self.database = Database()
        self.searchResults = self.database.Search(id)
        
        Label(self.window, text = self.searchResults[0][0],  width = 25).grid(pady = 5, column = 3, row = 1)
        Label(self.window, text = self.searchResults[0][1],  width = 25).grid(pady = 5, column = 2, row = 2)
        Label(self.window, text = self.searchResults[0][2],  width = 25).grid(pady = 5, column = 2, row = 3)
        Label(self.window, text = self.searchResults[0][3],  width = 25).grid(pady = 5, column = 2, row = 4)
        Label(self.window, text = self.searchResults[0][4],  width = 25).grid(pady = 5, column = 2, row = 5)
        
        Label(self.window, text = self.searchResults[0][0],  width = 25).grid(pady = 5, column = 3, row = 1)
        Label(self.window, text = self.searchResults[0][1],  width = 25).grid(pady = 5, column = 3, row = 2)

        #self.NameEntry = tkinter.Entry(self.window,  width = 25, textvariable = self.Medicine_name)
        self.mgEntry = Entry(self.window,  width = 25, textvariable = self.mg)
        self.stockEntry = Entry(self.window,  width = 25, textvariable = self.Avilable_stock)
        self.expEntry = Entry(self.window,  width = 25, textvariable = self.Expiry_date)
        

#        self.NameEntry.grid(pady = 5, column = 3, row = 2)
        self.mgEntry.grid(pady = 5, column = 3, row = 3)
        self.stockEntry.grid(pady = 5, column = 3, row = 4)
        self.expEntry.grid(pady = 5, column = 3, row = 5)
            

        Button(self.window, width = 20, text = "Update",activebackground="grey", command = self.Update).grid(pady = 15, padx = 5, column = 1, row = 14)
        Button(self.window, width = 20, text = "Close",activebackground="grey", command = self.window.destroy).grid(pady = 15, padx = 5, column = 3, row = 14)

        self.window.mainloop()

    def Update(self):
        self.database = Database()
        self.database.Update(self.mgEntry.get(),self.stockEntry.get(),self.expEntry.get(), self.id)
        messagebox.showinfo("Updated data", "Data updated successfully")

class DatabaseView:
    def __init__(self, data):
        self.databaseViewWindow = Tk()
        self.databaseViewWindow.title("Database View")

        Label(self.databaseViewWindow, text = "Database View Window",  width = 25).grid(pady = 5, column = 1, row = 1)

        self.databaseView = ttk.Treeview(self.databaseViewWindow)
        self.databaseView.grid(pady = 5, column = 1, row = 2)
        self.databaseView["show"] = "headings"
        self.databaseView["columns"] = ("id", "Medicine_name", "mg", "Avilable_stock", "Expiry_date")

        # Treeview column headings
        self.databaseView.heading("id", text = "Medicine ID")
        self.databaseView.heading("Medicine_name", text = "Medicine Name")
        self.databaseView.heading("mg", text = "Milli gram")
        self.databaseView.heading("Avilable_stock", text = "Avilable Stock")
        self.databaseView.heading("Expiry_date", text = "Expiry Date")
        

        # Treeview columns
        self.databaseView.column("id", width = 80)
        self.databaseView.column("Medicine_name", width = 200)
        self.databaseView.column("mg", width = 100)
        self.databaseView.column("Avilable_stock", width = 100)
        self.databaseView.column("Expiry_date", width = 100)
        

        for record in data:
            self.databaseView.insert('', 'end', values=(record))

        self.databaseViewWindow.mainloop()

class SearchDeleteWindow:
    def __init__(self, task):
        window = Tk()
        window.title(task + " data")

        self.id = StringVar()
        self.Medicine_name = StringVar()
        self.Avilable_stock = StringVar()
        self.heading = "Please enter Medicine ID to " + task

        Label(window, text = self.heading, width = 50).grid(pady = 20, row = 1)
        Label(window, text = "Medicine ID", width = 10).grid(pady = 5, row = 2)

        self.idEntry = Entry(window, width = 8, textvariable = self.id)
        self.idEntry.grid(pady = 5, row = 3)

        if (task == "Search"):
            Button(window, width = 20, text = task,activebackground="grey", command = self.Search).grid(pady = 15, padx = 5, column = 1, row = 14)
        elif (task == "Delete"):
            Button(window, width = 20, text = task,activebackground="grey", command = self.Delete).grid(pady = 15, padx = 5, column = 1, row = 14)
            
    def Search(self):
        self.database = Database()
        self.data = self.database.Search(self.idEntry.get())
        self.databaseView = DatabaseView(self.data)
    
    def Delete(self):
        self.database = Database()
        self.database.Delete(self.idEntry.get())
        messagebox.showinfo("Deleted data", "Data deleted successfully")
        
        
class HomePage:
    def __init__(self):
        self.homePageWindow = Tk()
        self.homePageWindow.title("Medicine Information System")
        
#        bg=PhotoImage(file="C:/Users/Admin/Downloads/27121-3-pills-image-thumb.png")
#         self.canvas1 = Canvas(self.homePageWindow).pack()
        Label(self.homePageWindow, text = "Home Page",  width = 100).grid(pady = 20, column = 1, row = 1)

        Button(self.homePageWindow, width = 20, text = "Insert",activebackground="grey", command = self.Insert).grid(pady = 15, column = 1, row = 2)
        Button(self.homePageWindow, width = 20, text = "Update ",activebackground="grey", command = self.Update).grid(pady = 15, column = 1, row = 3)
        Button(self.homePageWindow, width = 20, text = "Search medicine",activebackground="grey", command = self.Search).grid(pady = 15, column = 1, row = 4)
        Button(self.homePageWindow, width = 20, text = "Remove medicine",activebackground="grey", command = self.Delete).grid(pady = 15, column = 1, row = 5)
        Button(self.homePageWindow, width = 20, text = "Display all medicines",activebackground="grey", command = self.Display).grid(pady = 15, column = 1, row = 6)
        Button(self.homePageWindow, width = 20, text = "Exit",activebackground="grey", command = self.homePageWindow.destroy).grid(pady = 15, column = 1, row = 7)

        self.homePageWindow.mainloop()

    def Insert(self):
        self.insertWindow = InsertWindow()
    
    def Update(self):
        self.updateIDWindow = Tk()
        self.updateIDWindow.title("Update data")

        self.id = StringVar()

        Label(self.updateIDWindow, text = "Enter the ID to update", width = 50).grid(pady = 20, row = 1)

        self.idEntry = Entry(self.updateIDWindow, width = 8, textvariable = self.id)
        self.idEntry.grid(pady = 10, row = 2)
        
        Button(self.updateIDWindow, width = 20, text = "Update",activebackground="grey", command = self.updateID).grid(pady = 10, row = 3)

        self.updateIDWindow.mainloop()

    def updateID(self):
        self.updateWindow = UpdateWindow(self.idEntry.get())
        self.updateIDWindow.destroy()

    def Search(self):
        self.searchWindow = SearchDeleteWindow("Search")

    def Delete(self):
        self.deleteWindow = SearchDeleteWindow("Delete")

    def Display(self):
        self.database = Database()
        self.data = self.database.Display()
        self.displayWindow = DatabaseView(self.data)

homePage = HomePage()
```
